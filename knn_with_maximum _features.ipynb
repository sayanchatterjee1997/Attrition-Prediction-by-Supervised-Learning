```python
#Here I have dropped only the ID, EmployeeNumber, EmployeeCount and fit KNN on the dataset.
#The result shows that it gives the better accuracy than the other KNN where I deleted some unimportant features.

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
```

```python
file=pd.read_csv("train.csv")
#print(file.head())
```

```python
a1=file.drop("Attrition",axis=1)
a2=a1.drop("EmployeeCount",axis=1)
a3=a2.drop("EmployeeNumber",axis=1)
a4=a3.drop("ID",axis=1)
#print(a4.head())
```

```python
import matplotlib.pyplot as plt
df = pd.get_dummies(a4) 
print(df.corr())
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,file["Attrition"] ,test_size=0.2,random_state=20)
from sklearn.neighbors import KNeighborsClassifier

#Note that in the previous case where I extracted those important features, the best accuracy came  with k=8, here have tried with k=7,k=8,k=9,k=6 
#and found that the best accuracy comes at K=7. Hence to maintain neatness, I removed other iterations with k=8,9,6


model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)
b1=pd.read_csv("test.csv")
print(b1.head())
b2=b1.drop("EmployeeCount",axis=1)
b3=b2.drop("EmployeeNumber",axis=1)
b4=b3.drop("ID",axis=1)
df1=pd.get_dummies(b4)
y_predict1=model.predict(df1)
print(y_predict1)
df1=pd.DataFrame(y_predict1)
#df1.to_csv('C:/Users/SAYAN/Desktop/ML_assign2/KNN_7_check.csv')
```
